```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
data = pd.read_csv('forbes_billionaire_dataset.csv', error_bad_lines=False)
pd.set_option('precision', 9)
data.head(10)
```

```python
data.dtypes
```

```python
for i in range(len(data['age'])):
    if data['age'][i]<=0:
        data['age'][i]=50
data['age'].unique()
```

```python
data = data.fillna(0)
```

```python
data.describe()
```

```python
df=data[:10]
```

```python
plt.figure(figsize=(10,5))
plt.title('Top 10 Richest person in the World : 2018')
plt.xlabel('Names of Billionares')
plt.xticks(rotation=90)
plt.ylabel('Net Worth in Billions * 10')
plt.plot(df['name'], df['net_worth'], color='red', marker='*')
plt.show()
```

```python
df1=data['country_citizenship'].value_counts()
df1
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=90)
plt.title('Billionares per Country')
plt.xlabel('Name of Country')
plt.ylabel('Number of Billionares')
plt.plot(df1[:20], marker='*')
```

```python
df2 = data
ages_by_worth = df2['age'].apply(lambda x: np.float32(x) if x != '-' else np.nan)
plt.ylim(0,100)
plt.title('Age distribution vs Rank')
plt.xlabel('Rank')
plt.ylabel('Age')
plt.scatter(np.arange(len(ages_by_worth)), ages_by_worth)
```

```python
plt.title('Billionares Age Distribution')
sns.distplot(data['age'], color='black')
```

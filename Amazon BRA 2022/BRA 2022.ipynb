# Training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm


data = pd.read_csv('sample_csv.csv')
data.reset_index(drop = True, inplace=True)

data
```

```python
data['City'].unique()
```

```python
x= pd.get_dummies(data['City'])
data = data.drop(['City'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data = data.drop(['NaN\n'], axis=1)
```

```python
data['Type_of_order'].unique()
x= pd.get_dummies(data['Type_of_order'])
data = data.drop(['Type_of_order'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data['Weather conditions'].unique()
x= pd.get_dummies(data['Weather conditions'])
data = data.drop(['Weather conditions'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data['Vehicle_condition'].unique()
x= pd.get_dummies(data['Vehicle_condition'])
data = data.drop(['Vehicle_condition'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data['Road_traffic_density'].unique()
x= pd.get_dummies(data['Road_traffic_density'])
data = data.drop(['Road_traffic_density'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data['Festival'].unique()
x= pd.get_dummies(data['Festival'])
data = data.drop(['Festival'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data = data.drop(['Delivery_person_ID'], axis=1)
```

```python
data.head(10)
```

```python
data.dtypes
```

```python
data = data.drop(['NaN\n'], axis=1)
```

```python
data.dtypes
```

```python
data.head(10)
```

```python
data = data.drop(['Order_Date'], axis=1)
```

```python
data.head(10)
```

```python
data['latitude'] = data.apply(lambda row: row['Restaurant_latitude'] - row['Delivery_location_latitude'], axis=1)
```

```python
data.head(10)
```

```python
data['longitude'] = data.apply(lambda row: row['Restaurant_longitude'] - row['Delivery_location_longitude'], axis=1)
```

```python
data.head(10)
```

```python
data = data.drop(['Restaurant_longitude'], axis=1)
data = data.drop(['Delivery_location_longitude'], axis=1)
data = data.drop(['Restaurant_latitude'], axis=1)
data = data.drop(['Delivery_location_latitude'], axis=1)
```

```python
data.head(10)
```

```python
data.dtypes
```

```python
x= pd.get_dummies(data['Type_of_vehicle'])
data = data.drop(['Type_of_vehicle'], axis=1)
data = pd.concat([data, x], axis=1)
data.head(10)
```

```python
data.dtypes
```

```python
data['multiple_deliveries'].unique()
```

```python
data['multiple_deliveries'].replace('\n', '', regex=True, inplace=True)
```

```python
data['multiple_deliveries'] = data['multiple_deliveries'].astype(float)
```

```python
data.dtypes
```

```python
data['ID'].replace('\n', '', regex=True, inplace=True)
data['Delivery_person_Age'].replace('\n', '', regex=True, inplace=True)
data['Delivery_person_Ratings'].replace('\n', '', regex=True, inplace=True)
data['Time_Orderd'].replace('\n', '', regex=True, inplace=True)
data['Time_Order_picked'].replace('\n', '', regex=True, inplace=True)
```

```python
data.head(10)
```

```python
#data['ID'] = data['ID'].astype(float)
data['Delivery_person_Age'] = data['Delivery_person_Age'].astype(float)
data['Delivery_person_Ratings'] = data['Delivery_person_Ratings'].astype(float)
```

```python
data['Time_Orderd'].replace(':', '', regex=True, inplace=True)
```

```python
data['Time_Order_picked'].replace(':', '', regex=True, inplace=True)
```

```python
data['Time_Orderd'] = data['Time_Orderd'].astype(float)
data['Time_Order_picked'] = data['Time_Order_picked'].astype(float)
```

```python
data.dtypes
```

```python
data.head(10)
```

```python
data['Time_taken (min)'] = data['Time_taken (min)'].astype(int)
```

```python
data.head()
```

```python
features = data
features = features.drop(['ID', 'Time_taken (min)'],  axis=1)
target = data['Time_taken (min)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, random_state= 10)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(normalize=True)
X_train = X_train.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
regressor.fit(X_train, y_train)
X_test = X_test.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)
print("Predicted values:", regressor.predict(X_test))
```

```python
y_pred = regressor.predict(X_test)
y_pred.shape
```

```python
print("R^2 score for liner regression: ", 100*regressor.score(X_test, y_test))
```

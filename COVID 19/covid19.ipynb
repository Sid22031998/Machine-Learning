```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import glob, os
```

## Overall Analysis

```python
data = pd.concat(map(pd.read_csv, glob.glob(os.path.join('', "dataset\*.csv"))))
```

```python
data.fillna(0)
```

```python
data.dtypes
```

```python
data = data.drop(['Last_Update', 'Lat', 'Latitude', 'Long_', 'Longitude', 'Admin2', 'FIPS'], axis=1)
```

```python
data = data.drop(['Combined_Key'], axis=1)
```

```python
data = data.drop(['Last Update'], axis=1)
```

```python
data = data.fillna(0)
```

```python
data.head(10)
```

```python
data.dtypes
```

```python
data[7617:7620]
```

```python
data['Country_Region'].unique()
```

```python
plt.figure(figsize=(12,8))
plt.xlim(0, 300000, 50000)
plt.ylim(0, 160000, 40000)
plt.xlabel('Confirmed')
plt.title('Deaths Rate vs Recovery Rate')
plt.scatter(data['Confirmed'], data['Deaths'], color='red', label='Deaths')
plt.scatter(data['Confirmed'], data['Recovered'], color='blue', label='Recovered')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.regplot(data['Confirmed'], data['Active'])
```
